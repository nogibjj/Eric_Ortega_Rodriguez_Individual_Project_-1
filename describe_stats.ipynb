```python
# Importing Library 
import pandas as pd
```

```python
# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(url)

# Show part of the data
df.head()
```

```python
# Basic information about the dataset
df.info()
```

```python
# Descriptive statistics of numerical columns
df.describe()
```

```python
# Import visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Plot the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Survival rate by gender
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')
plt.title('Survival Count by Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Survival count by passenger class
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30, color='blue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Age distribution for survived vs. not survived
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 1]['Age'], kde=True, bins=30, color='green', label='Survived', alpha=0.6)
sns.histplot(df[df['Survived'] == 0]['Age'], kde=True, bins=30, color='red', label='Not Survived', alpha=0.6)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Survival rate by embarkation point
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set3')
plt.title('Survival Count by Embarkation Point')
plt.ylabel('Count')
plt.show()
```

```python
# Fare distribution by Pclass
plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'], kde=True, bins=30, color='purple')
plt.title('Fare Distribution of Titanic Passengers')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
